```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
```

```python
# Amazon URL
url = "https://www.amazon.in/s?k=laptop&crid=1CJSQ4Q5CIQSE&sprefix=%2Caps%2C372&ref=nb_sb_ss_recent_1_0_recent"
```

```python
# Headers
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}
```

```python
# Send request
response = requests.get(url, headers=headers)
```

```python
# Parse HTML
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Find all products
products = soup.find_all("div", {"data-component-type": "s-search-result"})
```

```python
# Empty list
all_data = []
```

```python
# Loop through products
for product in products:

    # Title
    title = "N/A"
    title_tag = product.find("h2")
    if title_tag:
        title = title_tag.get_text(strip=True)

    # Price
    price = "N/A"
    price_tag = product.find("span", class_="a-price-whole")
    if price_tag:
        price = price_tag.get_text(strip=True)

    # Rating
    rating = "N/A"
    rating_tag = product.find("span", class_="a-icon-alt")
    if rating_tag:
        rating = rating_tag.get_text(strip=True)

    # Image URL
    image = "N/A"
    image_tag = product.find("img", class_="s-image")
    if image_tag:
        image = image_tag.get("src")

    # Ad / Organic Result
    result_type = "Organic"

    sponsored_tag = product.find("span", string="Sponsored")

    if sponsored_tag:
        result_type = "Ad"

    # Store data
    all_data.append({
        "Title": title,
        "Price": price,
        "Rating": rating,
        "Image": image,
        "Result Type": result_type
    })
```

```python
# Create DataFrame
df = pd.DataFrame(all_data)
```

```python
# Timestamp for file name
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
```

```python
# CSV filename
file_name = f"amazon_laptop_data_{timestamp}.csv"
```

```python
# Save CSV
df.to_csv(file_name, index=False)

print("CSV File Saved:", file_name)
```
